```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import spotipy.util as util
```

```python
import os
client_id = os.environ.get("OAUTH_SPOTIFY_CLIENT_ID")
client_secret = os.environ.get("OAUTH_SPOTIFY_SECRET_ID")
```

```python
redirect_uri = "http://localhost:3000"
username = ""
```

```python
token = util.prompt_for_user_token(
    username,
    scope='user-read-recently-played',
    client_id=client_id,
    client_secret=client_secret,
    redirect_uri=redirect_uri
)
```

```python
spotify = spotipy.Spotify(auth=token)
```

```python
results = spotify.current_user_recently_played(limit=10)
```

```python
recent_tracks = []
for item in results['items']:
    track = item['track']
    track_dict = {
        'Nombre': track['name'],
        'Artista': track['artists'][0]['name'],
        'Album': track['album']['name'],
        'Fecha': item['played_at']
    }
    recent_tracks.append(track_dict)
```

```python
import json
results_json = json.dumps(results)
```

```python
df = pd.DataFrame(recent_tracks)
```
